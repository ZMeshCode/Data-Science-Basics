```python
# Drop rows with missing values
data.dropna(inplace=True)
```

```python
# Filter the relevant columns
relevant_columns = [
    'artist.id', 'artist.name', 'artist.familiarity', 'artist.hotttnesss',
    'song.id', 'song.title', 'song.year', 'song.duration', 'song.tempo',
    'song.key', 'song.loudness', 'song.mode', 'song.hotttnesss'
]
```

```python
data_filtered = data[relevant_columns]
data_filtered.head
```

### Pose a question

```python
Does the average hotttnesss score of songs change over time?
```

###  Visualization

```python
import matplotlib.pyplot as plt
# Filter out rows with unrealistic years ( before 1900)
data_filtered_valid_years = data_filtered[data_filtered['song.year'] >= 1900]

# Group data by year and calculate the average hotttnesss score for each year
average_hotttnesss_by_year_valid = data_filtered_valid_years.groupby('song.year')['song.hotttnesss'].mean().reset_index()

# Plotting the trend for valid years
plt.figure(figsize=(10, 6))
plt.plot(average_hotttnesss_by_year_valid['song.year'], average_hotttnesss_by_year_valid['song.hotttnesss'], marker='o')
plt.title('Average Song Hotttnesss Score Over Years (Valid Years Only)')
plt.xlabel('Year')
plt.ylabel('Average Hotttnesss Score')
plt.grid(True)
plt.show()
```

```python
#this data shows that around 1950 there was huge increase in average hotttness to songs propbably attributed to music being 
#more available to peeople. It also shows there has been a gradual increase to average hotttness since 1960s showing the more 
#recent song the higher average hotttness score
```

```python
import seaborn as sns

# Pivot the data to create a heatmap-friendly format
heatmap_data = data_filtered_valid_years.pivot_table(values='song.hotttnesss', index='song.year', aggfunc='mean')

# Create the heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt=".2f", linewidths=.5)
plt.title('Heatmap of Average Song Hotttnesss Score Over Years')
plt.xlabel('Year')
plt.ylabel('Average Hotttnesss Score')
plt.show()
```
