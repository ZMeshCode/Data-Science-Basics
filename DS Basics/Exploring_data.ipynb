### Exploring data

For each of the questions I proposed in Problem 2, I will perform an exploratory data analysis designed to address the question. I will produce two visualizations related to each question.I will describe what the visuals show and how they speak to my question of interest.

```python
import matplotlib.pyplot as plt

sns.set_style('whitegrid')
```

```python
# I used two bar charts to show the outcome distribution for different variables. This allows me to see what factors are more influential. 
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Animal Type', hue='Outcome Type')
plt.title('Outcome Types by Animal Type')
plt.xlabel('Animal Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Outcome Type')
plt.show
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Intake Condition', hue='Outcome Type')
plt.title('Outcome Types by Intake Condition')
plt.xlabel('Intake Condition')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Outcome Type')
plt.show()
# I used two bar charts to show the outcome distribution for different variables. This one is outcome from their intake condition 
```

```python
survival_counts = data.groupby(['Intake Condition', 'outcome_is_dead']).size().unstack(fill_value=0)
survival_counts_normalized = survival_counts.div(survival_counts.sum(axis=1), axis=0)

survival_counts_normalized.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Survival Rates by Intake Condition')
plt.xlabel('Intake Condition')
plt.ylabel('Proportion')
plt.legend(title='Outcome Status', labels=['Alive', 'Dead'])
plt.xticks(rotation=45)
plt.show() 
#This visualization is helpful in showing the proportion of survival rates with each intake condition category.
```

```python
# Recalculate 'Age at Intake' to ensure it's in the dataset this gives the aniaml of intake in days 
data['DOB'] = pd.to_datetime(data['DOB'], errors='coerce')
data['Intake Date'] = pd.to_datetime(data['Intake Date'], errors='coerce')
data['Age at Intake'] = (data['Intake Date'] - data['DOB']).dt.days

# Select relevant columns for correlation analysis
relevant_columns = ['Age at Intake', 'latitude', 'longitude', 'outcome_is_dead']
data_subset = data[relevant_columns].dropna()

# Calculate the correlation matrix
correlation_matrix = data_subset.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap of Relevant Variables')
plt.show()
#The heatmap shows the correlation coefficients between these variables, providing insight into their relationships. 
#For example, you can observe if there's a strong correlation between the age at intake and the survival status of the animals.
```
