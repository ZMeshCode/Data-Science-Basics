In this lab, I will analyze and visualize data from the Titanic disaster to explore two primary questions:

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

### Problem 1: Who were the Titanic passengers?

Let's take a look at the titanic passengers in terms of their age and recorded gender (labelled "sex" in the dataset). To examine this, create a violin plot visualization that shows a histogram of ages for each sex, broken down by whether the passenger survived.

```python
# set up survival status labels 
titanic['survived_str'] = titanic['survived'].map({0: 'Did Not Survive', 1: 'Survived'})

# Set the style to add gridlines
sns.set_style("whitegrid")

# violin plot
plt.figure(figsize=(12, 6))
ax = sns.violinplot(data=titanic, x='sex', y='age', hue='survived_str', split=True, palette={'Did Not Survive': 'red', 'Survived': 'green'})
plt.title('Distribution of Ages by Sex and Survival Status')
plt.xlabel('Sex')
plt.ylabel('Age')

# fix legend to have the right keys
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, ['Did Not Survive', 'Survived'], title='Survival Status', loc='upper left')

plt.show()
```

### Problem 2: A visualization of your own

```python
# Create a scatter plot to show the relationship between age, fare, and survival status
plt.figure(figsize=(14, 8))

# Use scatter plot with hue for survival status
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived_str', style='sex', s=100, alpha=0.7)

# Add title and labels
plt.title('Relationship Between Age, Fare, and Survival Status')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survival Status', loc='upper right')

plt.show()
#This scatter plot shows the relationship between passengers' age, fare paid,
#and their survival status on the Titanic. Each point represents a passenger, 
#with colors indicating whether they survived (green) or did not survive (red).  
#The plot reveals that younger passengers and those who paid higher fares had a higher likelihood of survival
```
