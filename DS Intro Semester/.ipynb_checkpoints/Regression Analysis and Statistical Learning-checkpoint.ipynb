```python
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
```

### Problem 1: Set up data

What is the relationship between the number of at-bats a team gets and the number of runs they score over the course of an entire baseball season? This is the question we will explore in this lab. 

#### (a) Import data

To begin, load the data found in the `GL2010.TXT`, `GL2011.TXT`, `GL2012.TXT`, `GL2013.TXT` into a single dataframe. You can find column names for the data in the `cnames.txt` file. You should ultimately have 9,720 rows of data.

```python
pd.set_option('display.max_columns', None) 
data_files = ['GL2010.TXT', 'GL2011.TXT', 'GL2012.TXT', 'GL2013.TXT']
column_file = ('cnames.txt')
columns = pd. read_csv('cnames.txt') 
```

```python
data = pd.concat([pd.read_csv(d, header=None) for d in data_files])
data.columns = columns.iloc[:, 0]
```

#### (b) Pull years

For this exercise, we need data for each team (30 in total) for each season in the dataset (4 seasons). Each season is designated by the year in which the games were played, which can be pulled from the `Date` variable. Start by creating a `Year` variable.

```python
# Create a Year variable from the Date column
data['Year'] = data['Date'].astype(str).str[:4]

data[['Date', 'Year']].tail()
```

#### (c) Aggregate data

Next, aggregate the scores and at-bats for each team across each season. Note that we do not make a distinction between home team and away team. You should have 120 records after aggregating (one for each team for each season).

```python
# Visitor data
visitor_data = data[['Visitor', 'Year', 'VisitorScore', 'Vat-bats']].rename(columns={
    'Visitor': 'Team',
    'VisitorScore': 'Score',
    'Vat-bats': 'At-bats'
})

# Home data
home_data = data[['Home', 'Year', 'HomeScore', 'Hat-bats']].rename(columns={
    'Home': 'Team',
    'HomeScore': 'Score',
    'Hat-bats': 'At-bats'
})
# Combine both
combined_data = pd.concat([visitor_data, home_data])
```

```python
# Aggregate scores and at-bats for each team across each season
aggregated_data = combined_data.groupby(['Team', 'Year']).agg({
    'Score': 'sum',
    'At-bats': 'sum'
}).reset_index()

aggregated_data.head()
```

### Problem 2: Regression

#### (a) Plot relationship

Plot the relationship between runs and at-bats. Does the relationship look linear? Describe the
relationship between these two variables.

```python
import matplotlib.pyplot as plt

# Set the style of the visualization
sns.set(style="whitegrid")

# Create a scatter plot with a regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='At-bats', y='Score', data=aggregated_data, ci=None, line_kws={"color": "red"})
plt.title('Relationship between At-Bats and Runs')
plt.xlabel('Total At-Bats')
plt.ylabel('Total Runs (Scores)')
plt.grid(True)
plt.show()
```

#### (b) Assessing relationship

If you knew a team’s at bats, would you be comfortable using a linear model to predict the number of
runs? Why or why not? Quantify the strength of the relationship with the correlation coefficient. Discuss what you find.

```python
# Calculate the correlation coefficient
correlation = aggregated_data['At-bats'].corr(aggregated_data['Score'])

correlation
```

```python
''' I calculated the correlation coefficient and found it to be 0.56, which is a moderate postive relationship. This relationship is not strong enough however 
for me to comfortable to use to predict the number of runs.  , relying solely on at-bats to predict runs would likely result in a model with significant
variance and potential inaccuracies.''' 
```

#### (c) Modeling

Use the `statsmodels` package to fit a simple linear model for runs as a function of at-bats. Ensure a constant (intercept) has been added to the regression. Write down the formula for the model, filling in estimated coefficient values.

```python
# Prepare the data
# Add a constant to the independent variable to include an intercept in the model
X = sm.add_constant(aggregated_data['At-bats'])
y = aggregated_data['Score']

# Fit the linear model
model = sm.OLS(y, X).fit()

# Output the summary of the model
model_summary = model.summary()
print(model_summary)

# Extract the coefficients
intercept, slope = model.params

# Write down the formula for the model
print(f"Formula: Runs = {intercept:.2f} + {slope:.2f} * At-bats")
```

#### (d) Interpretation

Describe in words the interpretation of $\beta$<sub>1</sub>.

```python
''' If we look at the beta the value is 0.55 which is the slope of the line.
A slope of 0.55 indicates that for each adiotnal bat you can expect 0.55 rruns '''
```

### Problem 3: Plotting

#### (a) Plot regression line

Make a plot of at-bats vs runs but include the fitted regression line. How well does the fitted line fit the data?

```python
import matplotlib.pyplot as plt


sns.set(style="whitegrid")

# Create a scatter plot with the regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(x='At-bats', y='Score', data=aggregated_data, color='blue', label='Data Points')
plt.plot(aggregated_data['At-bats'], intercept + slope * aggregated_data['At-bats'], color='red', label='Fitted Line')


plt.title('At-Bats vs Runs with Fitted Regression Line')
plt.xlabel('Total At-Bats')
plt.ylabel('Total Runs (Scores)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

#### (b) Plot residuals

Make a plot of the residuals versus at bats. Is there any apparent pattern in the residuals plot?

```python
# Calculate the residuals
aggregated_data['PredictedScore'] = intercept + slope * aggregated_data['At-bats']
aggregated_data['Residuals'] = aggregated_data['Score'] - aggregated_data['PredictedScore']

# Plot the residuals
plt.figure(figsize=(10, 6))
sns.scatterplot(x='At-bats', y='Residuals', data=aggregated_data, color='blue')
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuals vs. At-Bats')
plt.xlabel('Total At-Bats')
plt.ylabel('Residuals')
plt.grid(True)

# Show the plot
plt.show()
```
