```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# LeBron's stats for the last 5 seasons (2019-2023)
# Format: [Season, Age, Points per game, Rebounds per game, Assists per game]
data = [
    [2019, 34, 27.4, 8.5, 8.3],
    [2020, 35, 25.3, 7.8, 10.2],
    [2021, 36, 25.0, 7.7, 7.8],
    [2022, 37, 30.3, 8.2, 6.2],
    [2023, 38, 28.9, 8.3, 6.8]
]

X = np.array([row[1] for row in data]).reshape(-1, 1)  # Age as feature
y_points = np.array([row[2] for row in data])
y_rebounds = np.array([row[3] for row in data])
y_assists = np.array([row[4] for row in data])

# Create and fit the models
model_points = LinearRegression().fit(X, y_points)
model_rebounds = LinearRegression().fit(X, y_rebounds)
model_assists = LinearRegression().fit(X, y_assists)

# Project stats for next season (2024, age 39)
next_season_age = np.array([[39]])
projected_points = model_points.predict(next_season_age)[0]
projected_rebounds = model_rebounds.predict(next_season_age)[0]
projected_assists = model_assists.predict(next_season_age)[0]

print(f"Projected stats for LeBron James in 2024 (age 39):")
print(f"Points per game: {projected_points:.1f}")
print(f"Rebounds per game: {projected_rebounds:.1f}")
print(f"Assists per game: {projected_assists:.1f}")

# Plotting
plt.figure(figsize=(12, 8))
plt.scatter(X, y_points, color='blue', label='Actual Points')
plt.plot(X, model_points.predict(X), color='blue', linestyle='--', label='Points Trend')
plt.scatter(X, y_rebounds, color='red', label='Actual Rebounds')
plt.plot(X, model_rebounds.predict(X), color='red', linestyle='--', label='Rebounds Trend')
plt.scatter(X, y_assists, color='green', label='Actual Assists')
plt.plot(X, model_assists.predict(X), color='green', linestyle='--', label='Assists Trend')

plt.scatter(next_season_age, [projected_points], color='blue', marker='*', s=200)
plt.scatter(next_season_age, [projected_rebounds], color='red', marker='*', s=200)
plt.scatter(next_season_age, [projected_assists], color='green', marker='*', s=200)

plt.title("LeBron James Performance Projection")
plt.xlabel("Age")
plt.ylabel("Stats per Game")
plt.legend()
plt.grid(True)
plt.show()
```
